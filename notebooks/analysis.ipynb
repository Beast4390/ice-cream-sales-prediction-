# Ice Cream Sales Prediction - EDA

This notebook performs exploratory data analysis (EDA) on the ice cream dataset and produces visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('..\data\IceCreamData.csv')

# Basic overview
print('Shape:', df.shape)
display(df.head())
display(df.describe())
display(df.isna().sum())
```

```python
# Scatter plot of Temperature vs Revenue
plt.figure(figsize=(8, 5))
plt.scatter(df['Temperature'], df['Revenue'], alpha=0.7)
plt.title('Temperature vs Revenue')
plt.xlabel('Temperature (°C)')
plt.ylabel('Revenue')
plt.grid(True, alpha=0.2)
plt.show()
```

```python
# Quick model fit (Linear Regression) to inspect relationship
from sklearn.linear_model import LinearRegression

X = df[['Temperature']]
y = df['Revenue']

model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, alpha=0.5, label='Actual')
plt.plot(X, y_pred, color='orange', linewidth=2, label='Regression line')
plt.title('Linear Regression: Revenue vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()
```
